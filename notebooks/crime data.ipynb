## Export sample data for development

```python
import json
```

```python
def serialize_as_json(row):
    return {
        'latitude': row.Latitude,
        'longitude': row.Longitude,
        'description': row.PD_DESC,
        'offense': row.OFNS_DESC,
        'report_date': row.RPT_DT,
        'incident_date': row.CMPLNT_FR_DT,
        'incident_time': row.CMPLNT_FR_TM
    }

mock_server_response = json.dumps(list(bike_complaints.head(10000).apply(serialize_as_json, axis=1)))
mock_server_response
```

## Write data to postgres

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://')
```

```python
cleaned = bike_complaints
# cleaned = big_bike_complaints
```

```python
column_mapping = {
    'CMPLNT_NUM': 'complaint_number', 
    'PD_DESC': 'description',
    'OFNS_DESC': 'offense',
    'RPT_DT': 'report_date',
    'CMPLNT_FR_DT': 'incident_date',
    'CMPLNT_FR_TM': 'incident_time',
    'Latitude': 'latitude',
    'Longitude': 'longitude'
}

sql_df = cleaned[list(column_mapping.keys())]
sql_df = sql_df.rename(columns=column_mapping)
sql_df
```

```python
from sqlalchemy.types import String, Date, Time
```

```python
sql_df.to_sql('crime', engine, if_exists='append', index=False, dtype={
    'complaint_number': String,
    'description': String,
    'offense': String,
    'report_date': Date,
    'incident_date': Date,
    'incident_time': Time,
    'latitude': String,
    'longitude': String
})
```
