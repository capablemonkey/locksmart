## 2019 YTD complaints

```python
df = pd.read_csv("../data/nypd/NYPD_Complaint_Data_Current__Year_To_Date_.csv") 
```

```python
bike_complaints = df[df['PD_DESC'].notnull() & df['PD_DESC'].str.contains('BICYCLE')]
bike_complaints
```

```python
gdf = geopandas.GeoDataFrame(bike_complaints, geometry=geopandas.points_from_xy(bike_complaints.Longitude, bike_complaints.Latitude))
```

```python
gdf.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
```

```python
bike_complaints
```

```python
import folium
from folium import plugins
from folium.plugins import HeatMap
```

```python
m = folium.Map(location=(40.68679, -73.97851), zoom_start=10)

interval = bike_complaints[(bike_complaints.CMPLNT_TO_DT < '05/01/2019') & (bike_complaints.CMPLNT_TO_DT > '04/01/2019')]
coords = interval.apply(lambda r: (r["Latitude"], r['Longitude']), axis=1)
HeatMap(coords, radius=17).add_to(m)
m
```
